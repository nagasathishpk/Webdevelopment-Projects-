```python
import pandas as pd
import numpy as np
from matplotlib import *
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
d1= pd.read_csv("chennaihousepriceprediction.csv")
d1.head()
```

```python
d1.shape
```

```python
d1.groupby('AREA')['AREA'].agg('count')
```

```python
d2=d1.drop(['DIST_MAINROAD','SALE_COND','SALE_COND','UTILITY_AVAIL','MZZONE','COMMIS','PARK_FACIL'],axis='columns')
d2.head()
```

```python
d2.isnull().sum()
```

```python
d3=d2.dropna()
d3.isnull().sum()
```

```python
d3.shape
```

```python
d3['AREA'].unique()
```

```python
d3['INT_SQFT'].unique()
```

```python

area = d3.AREA.unique()
area
```

```python
for i in ['Ana Nagar','Ann Nagar']:
    d3['AREA']=d3['AREA'].replace(i,'Anna Nagar')
for i in ['Chrompt','Chrmpet','Chormpet']:
    d3['AREA']=d3['AREA'].replace(i,'Chrompet')
for i in ['Adyr']:
    d3['AREA']=d3['AREA'].replace(i,'Adyar')
for i in ['KKNagar']:
    d3['AREA']=d3['AREA'].replace(i,'KK Nagar')
for i in ['TNagar']:
    d3['AREA']=d3['AREA'].replace(i,'T Nagar')
for i in ['Velchery']:
    d3['AREA']=d3['AREA'].replace(i,'Velachery')
for i in ['Karapakam']:
    d3['AREA']=d3['AREA'].replace(i,'Karapakkam')
        
```

```python
d3.AREA.unique()
```

```python
d3.head()
```

```python
d3.AREA.unique()
```

```python
d3['BATHROOMS']=d3['N_BEDROOM']+d3['N_BATHROOM']
d3.head()
```

```python
d4=d3.drop(['N_BEDROOM','N_BATHROOM','STREET','PRT_ID','BUILDTYPE','QS_ROOMS','QS_BATHROOM','QS_BEDROOM'],axis='columns')
d4
```

```python
d5=d4.rename(columns={"N_ROOM":"BHK"})
d5
```

```python
d5['PRICE_PERSQFT']=d4['SALES_PRICE']/d4['INT_SQFT']
d5.head()
```

```python
d6=d5[~(d5['INT_SQFT']/d5['BHK']<300)]
```

```python
d6
```

```python
d7=d6.sort_values('AREA')
```

```python
d7
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
d7['AREA_NO']=le.fit_transform(d7['AREA'])
```

```python
lst = dict(zip(d7.AREA,d7.AREA_NO))
lst
```

```python
d7.head(20)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(d7[['AREA_NO','INT_SQFT','BHK','BATHROOMS']],d7[['SALES_PRICE']],test_size=0.2,random_state=10)
```

```python
x_test,y_test
```

```python
x_train, y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
lr_clf=KNeighborsRegressor()
```

```python
lr_clf.fit(d7[['AREA_NO','INT_SQFT','BHK','BATHROOMS']],d7[['SALES_PRICE']])
```

```python
lr_clf.score(d7[['AREA_NO','INT_SQFT','BHK','BATHROOMS']],d7[['SALES_PRICE']])
```

```python
lr_clf.predict([[1,1100,3,2]])
```
